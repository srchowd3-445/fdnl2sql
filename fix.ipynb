```python
import json
from collections import Counter

# Path to your file
path = "/mnt/data1/srchowd3/MAYO_AIM2/data/empty_gt_replaced_v4_fixed.json"

with open(path, "r", encoding="utf-8") as f:
    data = json.load(f)

# 1) Top-level key/value pair count per record
top_counts = [len(rec) for rec in data]

print(f"Records: {len(data)}")
print(f"Top-level key/value pairs per record:")
print(f"  min: {min(top_counts) if top_counts else 0}")
print(f"  max: {max(top_counts) if top_counts else 0}")
print(f"  avg: {sum(top_counts)/len(top_counts) if top_counts else 0:.2f}")

# 2) How often each top-level key appears
key_freq = Counter()
for rec in data:
    key_freq.update(rec.keys())

print("\nTop-level key frequencies:")
for k, v in key_freq.most_common():
    print(f"  {k}: {v}")

# 3) (Optional) Recursive key/value pair count including nested dicts/lists
def recursive_kv_count(obj):
    """
    Counts 'key-value pairs' across nested dicts.
    - For dict: counts len(dict) + nested counts inside values
    - For list: counts nested counts for each element
    - For primitives: 0
    """
    if isinstance(obj, dict):
        return len(obj) + sum(recursive_kv_count(v) for v in obj.values())
    if isinstance(obj, list):
        return sum(recursive_kv_count(x) for x in obj)
    return 0

recursive_counts = [recursive_kv_count(rec) for rec in data]

print("\nRecursive (nested) key/value pair counts per record:")
print(f"  min: {min(recursive_counts) if recursive_counts else 0}")
print(f"  max: {max(recursive_counts) if recursive_counts else 0}")
print(f"  avg: {sum(recursive_counts)/len(recursive_counts) if recursive_counts else 0:.2f}")
```
